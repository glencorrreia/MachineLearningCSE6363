```python
# Zoom in
# Zoom out
# Move left
# Move right
# Move down
# Move up
# Press
def to_gesture(model_score):
    model_score_name = []
    for i in model_score:
        if i == 0:
            model_score_name.append("ZoomIn")
        elif i==1:
            model_score_name.append("ZoomOut")
        elif i==2:
            model_score_name.append("MoveLeft")
        elif i==3:
            model_score_name.append("MoveRight")
        elif i==4:
            model_score_name.append("MoveDown")
        elif i==5:
            model_score_name.append("MoveUp")
        else:
            model_score_name.append("Press")
    return model_score_name

```
